```python
df.drop(columns='Unnamed: 0',inplace=True)
df
```

```python
### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

total_employees= df['Name'].count()

# number of employees in each team
team_count=df['Team'].value_counts()

# percentage splitting
percentage_split=(team_count/total_employees)*100

team_details=pd.DataFrame({'Team': team_count.index, 'Number of employees' : team_count.values, 'Percentage split' : percentage_split.values})
print('Team Details')
print(team_details)

# pie chart representation

plt.figure(figsize=(10,10))
plt.pie(team_count.values,labels=team_count.index,autopct='%1.1f%%')
plt.title('Percentage Splitting of Employees in Each Team')
plt.show()

# barchart representation

plt.bar(team_count.index,team_count.values,edgecolor='red')
# Set x-axis labels to be vertical
plt.xticks(rotation='vertical')
plt.title('Percentage Splitting of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Number of Employees')
plt.show
```

```python
### 2.Segregate the employees w.r.t different positions.

# number of employees in each position
position_count=df['Position'].value_counts()

position_details=pd.DataFrame({'Position': position_count.index,'Number of employees' : position_count.values})
print('Position Details')
print(position_details)

# piechart representation

plt.pie(position_count.values,labels=position_count.index,autopct='%1.1f%%')
plt.title('Number of Employees in Each Position')
plt.show()

# barchart representation

plt.bar(position_count.index,position_count.values,edgecolor='red')
plt.title('Number of Employees in Each Position')
plt.xlabel('Position')
plt.ylabel('Number of Employees')
plt.show()
```

```python
### 3.Find from which age group most of the employees belong to.

age_data=df['Age'].value_counts()

age_details=pd.DataFrame({'Age': age_data.index,'Number of employees': age_data.values})
print('Age Details')
print(age_details)

# Histogram representation

plt.hist(df['Age'],bins=[0,10,20,30,40,50],edgecolor='red')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()

print('Maximum number of employees belong to 20-30 age group')
```

```python
### 4.Find out under which team and position, spending in terms of salary is high.

max_salary_spending = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(max_salary_spending)
```

```python
### 5.Find if there is any correlation between age and salary , represent it visually.

# scatterplot representation
sns.scatterplot(x='Age',y='Salary',hue='Position',data=df)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age and Salary Correlation')
plt.show()

# Heatmap representation
new=df[['Age','Salary']]
corr=new.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

print('The correlation coefficient is :', df['Age'].corr(df['Salary']))
```
